```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def divided_diff(x, y):
    n = len(y)
    coef = np.zeros([n, n])
    coef[:,0] = y
    for j in range(1,n):
        for i in range(n-j):
            coef[i][j] = \
           (coef[i+1][j-1] - coef[i][j-1]) / (x[i+j]-x[i])        
    return coef

def newton_poly(coef, x_data, x):
    newton = coef[len(x_data) - 1 ]
    for k in range(1,len(x_data)):
        newton = coef[len(x_data) - 1 -k] + (x -x_data[len(x_data) - 1 -k])*newton
    return newton
```

```python
for i in range(4,16):
    x_k = []
    y_k = []
    a = []
    ln = []
    for k in range(0,i+1):
        x_k.append(1 + k/i)
        y_k.append(np.log(x_k[k]))
    a = divided_diff(x_k,y_k)[0, :]
    x_ = np.arange(x_k[0], x_k[len(x_k) - 1], .0001)
    
    for j in x_:
        ln.append(np.log(j))
        
    y_ = newton_poly(a, x_k, x_)
    plt.plot(x_, abs(y_ - ln))
    
    
```
